```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
from zplane import zplane
```

```python
NumeratorZcoefs = [1, -1, 1, -1]
DenominatorZcoefs = [1, 0.54048, -0.62519, -0.66354, 0.60317, 0.69341]

FreqResponse = signal.freqz(NumeratorZcoefs, DenominatorZcoefs, whole=True)
fig = plt.figure(figsize = [8, 6])
ax = fig.add_subplot(111)
ax.plot(FreqResponse[0], abs(np.array(FreqResponse[1])))
ax.set_xlim(0, 2*np.pi)
ax.set_xlabel("$\Omega$")
```

```python
z, p, k = signal.tf2zpk(NumeratorZcoefs, DenominatorZcoefs)
print(z, p, sep='\n')
# plt.figure(figsize=(8,8))
(z_plane, p_plane, k) = zplane(NumeratorZcoefs, DenominatorZcoefs, circlecolor='white')
print(z_plane, p_plane, k, sep='\n')
```
